## **Task 1 -Fill Mob No from Phone Number where Mob No is empty**

```python
data_main.isnull().sum()
```

```python
# Function to extract the last 10 digits of the phone number
def extract_last_10_digits(phone):
    return str(phone)[-10:]
```

```python
# Fill the null values in 'Mobile Number' with the last 10 digits of 'Phone Number'
data_main['Mobile Number'] = data_main.apply(
    lambda row: extract_last_10_digits(row['Phone Number']) if pd.isnull(row['Mobile Number']) or row['Mobile Number'] in ['None', 'nan'] else row['Mobile Number'],
    axis=1)

# Convert 'Mobile Number' to integer to remove any .0, then back to string
data_main['Mobile Number'] = data_main['Mobile Number'].apply(lambda x: str(int(float(x))))
```

## **Task 4 -Creating a new column Lead Stage**

```python
# a new column called lead stage 
# if eKYC has e12 OR e14 OR e15 then account opended AND Total Payment Amount is greater 100 then  
# else in proccess
```

```python
#data_merged['Total Payment Amount'].fillna(0).astype(int)
data_merged['Total Payment Amount'] = pd.to_numeric(data_merged['Total Payment Amount'], errors='coerce').fillna(0).astype(int)
data_merged.head()
```

```python
data_merged.dtypes
```

```python
data_merged.insert(5, 'Lead Stage', None)
data_merged.head()
```

```python
# Define the generate_lead_stage function
def generate_lead_stage(row):
    ekyc_stage_code = row['eKYC Stage Code']
    total_payment_amount = row['Total Payment Amount']
    
    if 'e12' in ekyc_stage_code or 'e14' in ekyc_stage_code or 'e15' in ekyc_stage_code:
        if total_payment_amount > 100:
            return '999-Account Opened'
        else:
            return 'L4-Account Opened'
    else:
        return 'L2-in process'
```

```python
data_merged['Lead Stage'] = data_merged.apply(generate_lead_stage, axis=1)
data_merged.head()
```

```python
# Making values in 'eKYC Stage Code' as null where values in 'Lead Stage' are 'L4-Account Opened', '999-Account Opened'
data_merged.loc[data_merged['Lead Stage'].isin(['L4-Account Opened', '999-Account Opened']), 'eKYC Stage Code'] = None
data_merged
```

## **Task 6 - Creating Agency wise Everything**

```python
agency_groups = data_merged.groupby('agency')
agency_dfs = {agency: group.drop(columns=['agency']) for agency, group in agency_groups}
```

```python
list_of_agency = []
list_of_agency_df = []
for agency, df in agency_dfs.items():
    list_of_agency.append(agency)
    list_of_agency_df.append(df)

print(list_of_agency)
#print(list_of_agency_df)
```

```python
list_of_agency_df[3]
```

```python
list_of_agency_df[0].columns
```

```python
pivoted_df = []
for each_df in list_of_agency_df:
    each_df_pivot = pd.pivot_table(each_df, index='Primary Source Campaign', columns='Lead Stage', values='Mobile Number', aggfunc='count', fill_value=0)

    # adding a row of TOTAL
    sum_row = each_df_pivot.sum()
    sum_row.name = 'TOTAL'
    each_df_pivot = pd.concat([each_df_pivot, pd.DataFrame(sum_row).T])

    # Reset the index to make 'Primary Source Campaign' a regular column
    each_df_pivot = each_df_pivot.reset_index()

    # Ensure all desired columns are present
    required_columns = ['999-Account Opened', 'L2-in process', 'L4-Account Opened']
    for col in required_columns:
        if col not in each_df_pivot.columns:
            each_df_pivot[col] = 0
    
    # Reorder the columns
    # each_df_pivot = each_df_pivot[['Primary Source Campaign'] + required_columns]

    # renaming pivot column index to Primary Source Campaign
    each_df_pivot.rename(columns={'index':'Primary Source Campaign'}, inplace=True)
    
    # merging Bitly Links to it ---- # Create a dictionary to map 'Primary Source Campaign' to 'Bitly Link'
    campaign_to_bitly = each_df.set_index('Primary Source Campaign')['Bitly Link'].to_dict()

    # Add 'Bitly Link' column to data_merged_pivot by mapping
    each_df_pivot['Bitly Link'] = each_df_pivot['Primary Source Campaign'].map(campaign_to_bitly)
    
    # re-arranging the columns as per the requirements 
    each_df_pivot = each_df_pivot[["Bitly Link", "Primary Source Campaign", "999-Account Opened", "L2-in process", "L4-Account Opened"]]

    # Adding a new column 'Grand Total' which is the sum of the specified columns
    each_df_pivot["Grand Total"] = each_df_pivot[["999-Account Opened", "L2-in process", "L4-Account Opened"]].sum(axis=1)

    # adding the pivoted df in the list
    pivoted_df.append(each_df_pivot)

    # removing the column 'Bitly Link' from each df
    each_df.drop(['Bitly Link', 'Total Payment Amount'], axis=1, inplace=True)

# agency_pivotdf_dict = agency_dfs.copy()
# agency_pivotdf_dict.update(dict(zip(agency_dfs.keys(), pivoted_df)))

# Create a dictionary to store the original and pivoted DataFrames
agency_pivotdf_dict = {}
for key, original_df, pivoted in zip(agency_dfs.keys(), list_of_agency_df, pivoted_df):
    agency_pivotdf_dict[key] = [original_df, pivoted]
    
agency_dfs.update(agency_pivotdf_dict)
```

```python
type(agency_pivotdf_dict['BS'])
```

```python
import zipfile
from io import BytesIO
# Dictionary to store in-memory Excel files
excel_files = {}

# Iterate through the agency_pivotdf_dict to create in-memory Excel files
for agency_name, dfs in agency_pivotdf_dict.items():
    original_df, pivoted_df = dfs
    
    # Create a BytesIO object to hold the Excel file in memory
    excel_buffer = BytesIO()
    
    # Create an Excel writer object and add the DataFrames as sheets
    print('creating excel files...')
    with pd.ExcelWriter(excel_buffer, engine='xlsxwriter') as writer:
        original_df.to_excel(writer, sheet_name='Raw Data', index=False)
        pivoted_df.to_excel(writer, sheet_name='Summary', index=False)
    
    # Save the BytesIO object to the dictionary
    excel_files[agency_name] = excel_buffer

# Create a zip file containing all the in-memory Excel files
print('creating zip...')
zip_file_path = "mstock-data.zip"
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
    for agency_name, excel_buffer in excel_files.items():
        # Seek to the beginning of the BytesIO object before writing
        excel_buffer.seek(0)
        # Write the BytesIO object to the zip file with the agency name as the file name
        zipf.writestr(f"{agency_name}.xlsx", excel_buffer.read())

print(f"All files have been zipped into {zip_file_path}")
```
